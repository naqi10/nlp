```python
import nltk
import spacy
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python
words = ["eating","eats","eat","ate","hearing","stemming","rafting","ability","meeting","adjustable"]

for word in words:
    print(word," | ", stemmer.stem(word))
```

```python
nlp = spacy.load("en_core_web_sm")
doc = nlp("eating eats eat ate hearing stemming rafting ability meeting adjustable better")
for token in doc:
    print(token," | ", token.lemma_)
```

```python
doc = nlp("Brian Carter flew to mars yesterday.he carried biryani masala with him")
for token in doc:
    print(token," | ", token.pos_)
```

```python
amazn = """The company has multiple subsidiaries, including Amazon Web Services, providing 
cloud computing; Zoox, a self-driving car division; Kuiper Systems, a satellite Internet provider;
and Amazon Lab126, a computer hardware R&D provider. Other subsidiaries include Ring, Twitch, IMDb, and Whole Foods Market. 
Its acquisition of Whole Foods in August 2017 for US$13.4 billion substantially increased its
market share and presence as a physical retailer.[8] Amazon also distributes a variety of downloadable and streaming content through
its Amazon Prime Video, MGM+, Amazon Music, Twitch, Audible and Wondery[9] units. It publishes books through
its publishing arm, Amazon Publishing, film and television content through Amazon MGM Studios,
including the Metro-Goldwyn-Mayer studio, which was acquired in March 2022, and owns Brilliance Audio and Audible,
which produce and distribute audiobooks, respectively."""
```

```python
doc = nlp(amazn)
for token in doc:
    print(token," | ", token.pos_," | ", spacy.explain(token.pos_))
```

```python
doc = nlp(amazn)
filtered_tokens = []
for token in doc:
    if token.pos_ not in ["SPACE", "X", "PUNCT"]:
        filtered_tokens.append(token)
```
